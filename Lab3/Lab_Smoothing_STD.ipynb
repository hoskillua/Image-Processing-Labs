```python
'''
(1) Your own median filter algorithm
'''
def median_smoothing(img, ws=3):
    smoothed_img = img.copy()
    
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            if i+ws//2 < img.shape[0] and i-ws//2 >= 0 and j+ws//2 < img.shape[1] and j-ws//2 >= 0:
                smoothed_img[i,j] = np.median(img[i-ws//2:i+ws//2+1,j-ws//2:j+ws//2+1])
    return smoothed_img


    
```

```python
'''
(2) skimage median filter
'''
img = io.imread('pout.tif')

noisy_img_sp = random_noise(img, mode='s&p', amount=0.05)

smoothed_img = median_smoothing(noisy_img_sp, ws=3)
smoothed_img_sk = median(noisy_img_sp)
show_images([img,noisy_img_sp, smoothed_img, smoothed_img_sk])




```

```python
'''
(3) Gaussion Filters
'''
show_images([img,noisy_img_sp])
show_images([gaussian(noisy_img_sp, sigma=0.2),gaussian(noisy_img_sp, sigma=1.7),gaussian(noisy_img_sp, sigma=4)])

```
